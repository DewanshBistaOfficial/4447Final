```python
# Read the CSV file
df1 = pd.read_csv("job.csv")
df1.head()
```

```python
df2 = pd.read_csv("ai.csv")
df2.head()
```

```python
#Only keep relevant columns
df1_cleaned = df1[['normalized_salary', 'work_type', 'title']]
df2_cleaned = df2
df2_cleaned.rename(columns={'Job titiles': 'title'}, inplace=True)
```

```python
df1_cleaned.head()
```

```python
df2_cleaned.head()
```

```python
#Drop null values
print(len(df1_cleaned))
print(len(df2_cleaned))
df1_cleaned = df1_cleaned.dropna()
df2_cleaned = df2_cleaned.dropna()
print(len(df1_cleaned))
print(len(df2_cleaned))
```

# Write an alogrithm to map job titles

```python
def replace_titles(input):
    for str in df2_cleaned["Title"]:
        if str in input:
            return str
    return None
```

```python
df1_cleaned["title1"] = df1_cleaned["title"].apply(replace_titles)
```

```python
df1_cleaned
```

### Clean null values

```python
print(len(df1_cleaned))
df1_cleaned = df1_cleaned.dropna()
print(len(df1_cleaned))
```

## Get average salary per title and work type

```python
df1_averaged = df1_cleaned.groupby(['title1', 'work_type'])['normalized_salary'].mean().reset_index(name='average_normalized_salary')
```

```python
df1_averaged.head()
```

# Merge both datasets

```python
merged_df = pd.merge(df1_averaged, df2_cleaned, left_on='title1', right_on='Title', how='left')
```

```python
merged_df = merged_df.rename(columns={'title1': 'title'})
merged_df.head()
```
